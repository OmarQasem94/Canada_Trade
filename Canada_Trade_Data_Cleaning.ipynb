```python
# Importing dependencied
import pandas as pd
import numpy as np
```

```python
# Creating pandas dataframe
Can_Trade = pd.read_csv('12100119.csv', index_col=False)

# Viewing dataframe
Can_Trade
```

```python
# Cleaning the dataframe

# Removing unimportant columns
Canada_Trade_df = Can_Trade.drop(columns=['DGUID', 'UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS', 'COORDINATE'])

# Removing specific rows that will lead to double counting
Canada_Trade_df = Canada_Trade_df.drop(Canada_Trade_df[(Canada_Trade_df['GEO']=='Canada') & (Canada_Trade_df['Trade']!='Re-export')].index)
Canada_Trade_df = Canada_Trade_df.drop(Canada_Trade_df[(Canada_Trade_df['Principal trading partners']=='All countries')].index)
Canada_Trade_df = Canada_Trade_df.drop(Canada_Trade_df[(Canada_Trade_df['North American Product Classification System (NAPCS)']=='Total of all merchandise')].index)

# Multiplying the VALUE column by 1000 to correct the values to the exact dollar amount
Canada_Trade_df['VALUE'] = Canada_Trade_df['VALUE']*1000

# Inserting commas to the VALUE column and converting it back to 
Canada_Trade_df['VALUE'] = Canada_Trade_df.apply(lambda x: "{:,}".format(x['VALUE']), axis=1)
Canada_Trade_df
```
